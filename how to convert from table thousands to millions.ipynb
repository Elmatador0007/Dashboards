```python
import numpy as np
import pandas as pd
```

```python
smo = pd.read_csv('Downloads/smoke.csv')
smo
```

```python
print(smo['Smoking mortality'].unique())
```

```python
smo['Smoking mortality'] = smo['Smoking mortality'].astype(str)

# Rimuovi i punti dalle stringhe nella colonna 'Smoking mortality' e converti in float
smo['Smoking mortality'] = smo['Smoking mortality'].str.replace('.', '').astype(float)

# Dividi ogni valore per un milione e moltiplica per mille
smo['Smoking mortality'] = smo['Smoking mortality'] / 1e6 * 1e3

# Stampa il dataframe smo
smo
```

```python
def truncate_decimal(number):
    return int(number)

# Applica la funzione alla colonna 'Smoking mortality'
smo['Smoking mortality'] = smo['Smoking mortality'].apply(lambda x: truncate_decimal(x))

# Stampa il DataFrame per verificare le modifiche
```

```python
smo.to_csv('smo1_data.csv', index=False)
```
